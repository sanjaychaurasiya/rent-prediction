# Artifical Neural Network for Housing Rent Dataset 


## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
tf.__version__
```

## Importing the dataset

```python
dataset = pd.read_csv("HousingRentDataset.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Encoding the dataset

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[('encode', OneHotEncoder(), [3, 4, 5])], remainder='passthrough')
X = transformer.fit_transform(X)
```

```python
print(X)
```

## Spliting the dataset into train and test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train.shape)
```

```python
print(y_test)
```

## Feature Scaling

```python
# Reshaping the y data.
y_train = y_train.reshape(len(y_train), 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```python
print(X_train)
```

```python
print(y_train)
```

# Part-2 Building the ANN

## Initializing the ANN


```python
ann = tf.keras.models.Sequential()
```

## Adding the Input and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

## Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

## Adding third hidden layer.

```python
# ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

## Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1))
```

# Part-3 Training the ANN

## Compiling the ANN

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

## Training the ANN model on the train dataset

```python
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

## Predicting the result of the test set.


```python
y_pred = sc_y.inverse_transform(ann.predict(sc_X.transform(X_test)))
```

```python
np.set_printoptions(precision=2)
np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1)
```

## Testing the accuracy of the model.

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
